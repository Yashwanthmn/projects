```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing datasets
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table('users.dat', sep='::',header=None,names=unames)
rnames = ['user_id', 'movie_id', 'rating', 'timestamp'] 
ratings = pd.read_table('ratings.dat', sep='::', header=None,names=rnames)
mnames = ['movie_id', 'title', 'genres'] 
movies = pd.read_table('movies.dat', sep='::', header=None,names=mnames) 
```

```python
 #creating a master dataset
data = pd.merge(pd.merge(ratings, users), movies)
```

```python
val = movies.genres.str.split("|")

res_col = []
for v in val:
    for i in v:
        if i not in res_col:
            res_col.append(i)

res_col.append("gender")
res_col.append("age")
res_col.append("rating")

df = pd.DataFrame(columns=res_col)

```

```python
#listing all unique genres
res_col
```

```python
res = data.merge(movies['movie_id'])[['genres',"rating","gender", "age"]]

for index, row in movies.head(20000).iterrows():
    tmp = row.genres.split("|") 
    
    for i in tmp:
       # print(i)
        df.loc[index,i] = 1
        df.loc[index,"gender"] = res.loc[index,"gender"]
        df.loc[index,"age"] = res.loc[index,"age"]
        df.loc[index,"rating"] = res.loc[index,"rating"]
        df.loc[index,df.columns[~df.columns.isin(tmp+["gender","rating","age"])]] = 0

df.head()
```

```python
gender = pd.get_dummies(df['gender'],drop_first=True)
df=df.drop('gender',axis=1)
gender.head()
```

```python
df = pd.concat([df,gender], axis=1)
```

```python
df=df.drop('M',axis=1)
```

```python
#determine the feature aaffecting ratings of particular movie
data_ind = df.drop("rating",axis=1)
print(data_ind.head())
data_dep = df["rating"]
print(data_dep.head())
```

```python
#Split the data into 80% training dataset and 20% test dataset.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data_ind,data_dep,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
independent_scaler = StandardScaler()
X_train = independent_scaler.fit_transform(X_train)
X_test = independent_scaler.fit_transform(X_test)
```

```python
#Linear regression 
from sklearn.linear_model import LinearRegression
linearRegModel = LinearRegression(n_jobs=-1)
linearRegModel.fit(X_train,y_train)

#prediction
y_pred = linearRegModel.predict(X_test)
y_pred
```
