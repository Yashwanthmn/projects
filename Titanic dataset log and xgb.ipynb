```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('train.csv')
```

```python
test= pd.read_csv('test.csv')
```

```python
data.isnull().sum()
```

```python
Embarked=pd.get_dummies(data['Embarked'])
Sex=pd.get_dummies(data['Sex'])
SibSp=pd.get_dummies(data['SibSp'])
Pclass=pd.get_dummies(data['Pclass'])
```

```python
data=pd.concat([data,Sex,Embarked,Pclass,SibSp],axis=1)
```

```python
data.head(5)
```

```python
data.drop(['Sex','Embarked','Name','Cabin','Ticket','Parch','SibSp',],axis=1,inplace=True)
```

```python
data.drop(['PassengerId'],axis=1,inplace=True)
data.head(5)
```

```python
test.head()
```

```python
Embarked=pd.get_dummies(test['Embarked'],)
Pclass=pd.get_dummies(test['Pclass'])
Sex=pd.get_dummies(test['Sex'])
SibSp=pd.get_dummies(test['SibSp'])
```

```python
test=pd.concat([test,Sex,Embarked,Pclass,SibSp],axis=1)
```

```python
test.dropna(inplace=True)
for col in test.columns: 
    print(col)
```

```python
data.dropna(inplace=True)
for col in data.columns: 
    print(col)
```

```python
test.drop(['Sex','Embarked','Name','Cabin','Ticket','Parch','SibSp'],axis=1,inplace=True)
```

```python
test.drop(['PassengerId'],axis=1,inplace=True)
test.head()
```

```python
X_train = data.drop("Survived",axis=1)
print(X_train.head())
y_train = data["Survived"]
print(y_train.head())
X_test = test
print(X_test.head())
y_test = pd.read_csv('gender_submission.csv')
y_test=y_test["Survived"]
print(y_test.head())
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
my_model = LogisticRegression()
my_model.fit(X_train,y_train)
```

```python
predictions = my_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
prediction = [round(value) for value in predictions]
accuracy = accuracy_score(y_test, prediction)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
from xgboost import XGBClassifier
my_xgb = XGBClassifier()
my_xgb.fit(X_train,y_train, verbose=False)
```

```python
xgb_predictions = my_xgb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

prediction = [round(value) for value in xgb_predictions]
accuracy = accuracy_score(y_test, prediction)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
data1 = pd.read_csv('gender_submission.csv')
final_data = ('PassengerId','Survived')
final_data('PassengerId') = data1('PassengerId')
final_data('Survived') = predictions
```

```python
final_data.to_csv(r'submission.csv',index=False)
```
